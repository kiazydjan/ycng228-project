```python
def get_ticker_data(ticker=None, lag=None):
    if ticker==None:
        return f'No ticker provided to get_ticker_data function'
    else:
        if lag is None:
            df = si.get_data(ticker,index_as_date=False)
        else:
            df = si.get_data(ticker,index_as_date=False).tail(lag)
        return df
    
    
```

```python
display(get_ticker_data('AAPL',30))
```

```python
df1 = get_ticker_data('AAPL',30)
```

```python
df2 = df1['close'].fillna(method = 'bfill').reset_index().drop(columns='index').T
display(df2)
```
