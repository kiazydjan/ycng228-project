```python
def get_ticker_data(ticker=None, lag=None):
    if ticker is None:
        return f'No ticker provided to get_ticker_data function'
    else:
        if lag is None:
            df = si.get_data(ticker, index_as_date=False)
        else:
            df = si.get_data(ticker , index_as_date=False).tail(lag)
        return df
    
def create_prediction_input(df):
    x_predict = df['close'].fillna(method='bfill').reset_index().drop(columns='index').diff().dropna().T

    return x_predict
```

```python
import joblib
```

```python
model = joblib.load('mdl.sav')
```

```python
y_pred = model.predict(create_prediction_input(get_ticker_data('AAPL',31)))
print(y_pred)
```

```python
df_tst = get_ticker_data('AAPL',31)
```

```python
x_predict = df_tst['close'].fillna(method='bfill').reset_index().drop(columns='index').diff().dropna().T
display(x_predict)
```

```python
results = []
for x in si.tickers_sp500():
    y_pred = model.predict(create_prediction_input(get_ticker_data(x,31)))
    results.append([x,y_pred[0]])
```

```python
display(results)
```
